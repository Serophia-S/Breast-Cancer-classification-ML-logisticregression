IMPORTING THE LIBRARIES

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and processing

```python
# loading the data from sklearn
from sklearn.datasets import load_breast_cancer
breast_cancer_dataset=load_breast_cancer()
print(breast_cancer_dataset)
```

```python
#loading the data to a dataframe
df=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
#print first 5 rows of dataset
df.head()
```

```python
#including target column of the dataset to the dataframe
df['diagnosis']=breast_cancer_dataset.target
#print last 5 rows of dataframe
df.tail()
```

```python
#Number of rows and columns
df.shape
```

```python
#getting info about the dataset
df.info()
```

```python
#checking for missing values
df.isnull()
```

```python
df.isnull().sum()
```

```python
#statistical info about the data
df.describe()
```

```python
#No of benign and malignant cases
df['diagnosis'].value_counts()
```

**1--benign ,0--malignant**

```python
df.groupby('diagnosis').mean()
```

From the above output, it is clear that malignant cases have values greater than the benign values.

```python
#splitting features(x) and target(y)
x=df.drop(columns='diagnosis',axis=1)
#if a column is dropped,then axis should be 1
y=df['diagnosis']
```

```python
print(x)
```

```python
print(y)
```

**splitting the data into training data and testing**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

**Model training**


```python
#linear regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
model = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model.fit(x_train, y_train)
predictions = model.predict(x_test)
plt.scatter(y_test, predictions, label='Data Points')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--', label='Fitted Line')
plt.xlabel("Actual Diagnosis")
plt.ylabel("Predicted Diagnosis")
plt.title("Actual vs Predicted Diagnosis with Fitted Line (Linear Regression)")
plt.legend()
plt.show()
```

```python
#svm
from sklearn.svm import SVR
import matplotlib.pyplot as plt


# Create and train the SVR model
model = SVR(kernel='linear')
model.fit(x_train, y_train)

# Make predictions
predictions = model.predict(x_test)

# Plot the graph
plt.scatter(y_test, predictions, label='Data Points')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--', label='Fitted Line')
plt.xlabel("Actual Diagnosis")
plt.ylabel("Predicted Diagnosis")
plt.title("Actual vs Predicted diagnosis with Fitted Line (SVM)")
plt.legend()
plt.show()
```

```python

from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

# Initialize the decision tree regressor
model = DecisionTreeRegressor(random_state=42)

# Fit the model to the training data
model.fit(x_train, y_train)

# Predict on the test data
predictions = model.predict(x_test)

# Plot the graph
plt.scatter(y_test, predictions, label='Data Points')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--', label='Fitted Line')
plt.xlabel("Actual diagnosis")
plt.ylabel("Predicted diagnosis")
plt.title("Actual vs Predicted diagnosis with Fitted Line (Decision Tree)")
plt.legend()
plt.show()

```

**Logistic regression**

```python
model=LogisticRegression()
```

```python
#training the logistic regression model using training data
model.fit(x_train,y_train)
```

**Model Evaluation**


**Accuracy Score**

```python
#accuracy on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(y_train,x_train_prediction)
```

```python
print("Accuracy of training data=",training_data_accuracy)
```

```python
#accuracy on test data
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(y_test,x_test_prediction)
```

```python
print("accuracy of testing data=",testing_data_accuracy)
```

The main reason to find the accuracy of test and train data is that sometimes model will be overfit that is the training accuracy will be high whereas test accuracy will be low.
